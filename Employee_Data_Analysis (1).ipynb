```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

# Load the dataset using the specified file path
file_path = r"C:\Users\ARUNKRISHNA V\Downloads\myexcel - myexcel.csv.csv"
df = pd.read_csv(file_path)

# Print DataFrame information to check column names and data types
print(df.info())

# Print the first few rows of the DataFrame to inspect the data
print(df.head())

# Replace missing data in the "height" column with random numbers between 150 and 180
df['Height'] = df['Height'].apply(lambda x: random.randint(150, 180) if pd.isnull(x) else x)

# Distribution across teams
team_distribution = df['Team'].value_counts()
team_percentage = (team_distribution / len(df)) * 100

# Plotting
plt.figure(figsize=(8, 6))
team_percentage.plot(kind='bar', color='skyblue')
plt.title('Employee Distribution Across Teams')
plt.xlabel('Teams')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.show()

# Segregation by Positions
position_counts = df['Position'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
position_counts.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('Set3'))
plt.title('Employee Segregation by Positions')
plt.ylabel('')
plt.show()

# Age Distribution
sns.histplot(data=df, x='Age', bins=10, kde=True)
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Salary Expenditure by Team and Position
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
team_salary.plot(kind='bar', color='salmon')
plt.title('Salary Expenditure by Team')
plt.xlabel('Teams')
plt.ylabel('Salary Sum')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
position_salary.plot(kind='bar', color='lightgreen')
plt.title('Salary Expenditure by Position')
plt.xlabel('Positions')
plt.ylabel('Salary Sum')
plt.xticks(rotation=45)
plt.show()

# Correlation Between Age and Salary
plt.figure(figsize=(30, 6))
sns.scatterplot(data=df, x='Age', y='Salary', hue='Team')
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend(title='Teams')
plt.show()
```
